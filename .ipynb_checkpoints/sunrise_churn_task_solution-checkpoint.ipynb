### Data preparation

I use the MinMaxScaler class available in the Scikit-learn library which simply applies to all x observations the transformation $(x-x_{min})/(x_{max}-x_{min})$.  

Moreover, note that the normalization is "fit" on the training set (min and max values of each feature are derived from the training set) and then applied also to the test set. 


```python
to_normalize = ['tenure','MonthlyCharges','diffInTotalCharges']
```

### Churn modeling

How the model will be used in practice? This kind of model will be used to check if a given customer is likely to churn in the near future. If the model predicts a customer churn then the company can target the specific customer of category of similar customers and adopt a predifined strategy to increase loyalty. The model can also provide insights on the features importance so that the company can understand where to invest to improve its services. 

Given the high degree of non-linearity of the problem I decided to implement a random forest classifier, which will be also employed to get an overview and a ranking of the features based on their importance, and a slightly more complex neural network.

I firstly run these classifiers with default parameters and then I go ahead with hyperparameters optimization.

When working with imbalanced datasets or with models where it is more important to correctly predict an event (e.g. churn occurrency) rather than the entire populaiton, the confusion matrix provides a deeper picture of the model performance compared to the simple accuracy. 

Accuracy expresses the fraction (or count) of correct predictions and mathematically is represented as follows:
$$
Accuracy(y_{true}, y_{pred}) = \frac{1}{n}\sum_{i = 0}^{n-1}1(y_{true,i} = y_{pred,i}) 
$$

Confusion matrix in sklearn is represented as depicted in picture:

<img src="https://static.packt-cdn.com/products/9781838555078/graphics/C13314_06_05.jpg" width="400"/>

Where TP, TN, FP, FN are respectively representing the count of predicted true positives, true negatives, false positives, and false negatives. 

Other useful metrics to consider are:

$$
Precision = \frac{TP}{TP+FP}
$$

$$
Recall = \frac{TP}{TP+FN}
$$

$$
F1 = 2\frac{Precision * recall}{Precision + Recall}
$$

Below I create a function for the estimation of all these KPIs and metrics. As mentioned above I will not considere accuracy because given the problem at hand it is much more important to correctly predict True Positives rather than the entire dataset. Between Precision and Recall I prefer Recall because it is more important to identify a customer who left the company but has been predicted as loyal (FN) rather than having a model which focuses on the distinction between TP and FP. In other words (and also economically speaking) it is better to focus on reducing the FN because this implies a correct estimation of economic loss rather than focusing on FP which only implies an underestimation of the total capital gain (less riskier for the company). 

Finally, F1_score offers a good compromise between precision and recall. 

To summarize, the most important KPI is Recall followed by F1_score and Precision. Accuracy is not important as the other metrics. 

NOTE: The *ml_func* created below can be easily adapted to work and fit other sklearn model object classifiers such as logistic regression, K-nearest neighbors classifier, decision trees etc...  

### Random Forest model

```python
def get_performance(y_true, y_pred):
    """
    This function calculates and prints accuracy score, precision, recall, F1 score and displays the confusion matrix
    
    Parameters:
    ----------
    y_true : numpy.ndarray
        corresponding observed labels.
    y_pred : numpy.ndarray
        corresponding predicted labels.
    
    Returns:
    ----------
    None
    """
    accuracy = accuracy_score(y_true, y_pred, normalize=True)
    print("accuracy: ",accuracy)
    precision = precision_score(y_true, y_pred)
    print("precision: ",precision)
    recall = recall_score(y_true, y_pred)                            
    print("recall: ",recall)
    f1 = f1_score(y_true, y_pred)
    print("f1_score: ",f1)
    
    # confusion matrix
    conf_matrix = confusion_matrix(y_true,y_pred)

    # confusion matrix visualization
    f, ax = plt.subplots(figsize = (5,5))
    sns.heatmap(conf_matrix, annot = True, linewidths = 0.5, color = "red", fmt = ".0f", ax=ax)
    plt.xlabel("y_predicted")
    plt.ylabel("y_true")
    plt.title("Confusion Matrix")
    plt.show()
```

```python
def ml_func(algorithm, X_train, y_train, X_test, y_test, param_dictionary=None, get_model=False):
    """
    Trains, predicts, and evaluates performance of the ML algorithm passed.

    Parameters
    ----------
    algorithm : sklearn object.
    X_train : numpy.ndarray
        train dataset.
    y_train : numpy.ndarray
        corresponding train labels.
    X_test : numpy.ndarray
        test dataset.
    y_test : numpy.ndarray
        corresponding test labels.
    param_dictionary: python dictionary
        dictionary with key-value pairs of parameters and their associated values.
    get_model: bool
        set to True if the model has to return. Default value, False
    Returns
    -------
    model: sklearn model object.
        Fitted model object

    """
    
    
    # train and fit model    
    if param_dictionary == None:
        model = algorithm()
    else:
        model = algorithm(**param_dictionary)
        
    model.fit(X_train, y_train)

    # predict
    train_preds = model.predict(X_train)
    test_preds = model.predict(X_test)

    print("------------------------")
    print(str(algorithm))
    print("------------------------")

    # evaluate all the efficiency metrics (accuracy, confusion matrix, precision, recall, F1 score)
    print('Train set performance \n')
    get_performance(y_train, train_preds) # train performance
    
    print('Test set performance \n')
    get_performance(y_test, test_preds) # test performance
    
    if get_model:
        return model
```

```python
# list of all classifiers that I will run for base models
algorithms = [RandomForestClassifier]

# running each model and print accuracy scores
print('Training without any optimization')
for algorithm in algorithms:
    ml_func(algorithm, X_train_scaled, y_train, X_test_scaled, y_test)
```

I now want to optimize the RF model using grid search of multiple parameter values. Grid search allows explore the hyperparameters space and to figure out the best parameters to get the most accurate result. 

I create a function for grid search named best_param. It will take parameter values and the
classifer and print our the best parameter combinations.

```python
def grid_src(classifier, param_grid, X_train, y_train):
    """
    This function implement a grid searching algorithm to define the best
    parameter set of a given classifier over a param_grid.

    Parameters
    ----------
    classifier : sklearn model object
        The model that has to be optimized.
    param_grid : dictionary
        dictionary whose keys are model parameters and values the set of
        parameters to investigate.

    Returns
    -------
    dictionary
        Returns a dictionary with the best parameters found.
    """
    
    # instantiate the gridsearch class using 5-fold cross validation, all processors (to speed up the training)
    # and use recall as scoring metric.
    grid_search = GridSearchCV(classifier, param_grid, cv=5, n_jobs=-1, verbose=2, scoring='recall')

    # train the tuned random forest
    grid_search.fit(X_train, y_train)

    # print best estimator parameters found during the grid search
    print((str(classifier) + 'Best Parameters'))
    print("------------------------")
    print(f"Best params: \n{grid_search.best_params_}")
    print(f"Recall of the best model: \n{grid_search.best_score_}")
    return grid_search.best_params_

```

```python
# Grid Search for best parameters of RandomForestClassifier
param_grid_rf = {'n_estimators': [100, 500, 1000], # number of trees in the forest
                 'criterion': ['gini','entropy'], # The function to measure the quality of a split.
                 'max_depth': [10, 20, 28], # The maximum depth of the tree
                 'min_samples_split': [2, 4], # The minimum number of samples required to split an internal node
                 'max_features': ['sqrt'] # The number of features to consider when looking for the best split
                }
      
rf_params = grid_src(RandomForestClassifier(), param_grid_rf, X_train_scaled, y_train)
```

Fit now the model using the best parset found and check recall on the test set:

```python
# list of all classifiers that I will run for base models
algorithms = [RandomForestClassifier]

# running each model and print accuracy scores
print('Training using best parameters found')
for algorithm in algorithms:
    best_rf = ml_func(algorithm, X_train_scaled, y_train, X_test_scaled, y_test, 
                    param_dictionary=rf_params, get_model=True)
```

### Neural Network

Despite the dataset is not particularly big, I chose to train and optimize a deep neural network. The architecture chosen is a multilayer perceptron network (MPN), i.e. a series of fully connected layers in which the last layer activation is a sigmoid function to allow the NN to perform the binary classification.

Firstly, I train a MPN with all hyperparameters chosen, then I optimize it. The only hyperparameter I am immediately interested in finding a reasonable value is the learning rate and for which I introduce a callback to test for each epoch a different value. 

Since metrics have been removed from Keras core, I implement a backend to get at the end of each epoch the KPIs I am interested (Recall, F1score, Precision).

```python
def recall_m(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    recall = true_positives / (possible_positives + K.epsilon())
    return recall

def precision_m(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    return precision

def f1_m(y_true, y_pred):
    precision = precision_m(y_true, y_pred)
    recall = recall_m(y_true, y_pred)
    return 2*((precision*recall)/(precision+recall+K.epsilon()))
```

```python
def dnn_func(X_train, y_train, X_test, y_test, epochs=100, batch_size=16, l_r_callback=False):
    """
    Trains, predicts, and evaluates performance of a four layers dnn.

    Parameters
    ----------
    X_train : numpy.ndarray
        train dataset.
    y_train : numpy.ndarray
        corresponding train labels.
    X_test : numpy.ndarray
        test dataset.
    y_test : numpy.ndarray
        corresponding test labels.
    epochs : int32
        number of times the dnn sees the entire trainset
    batch_size: int32
        number of samples to load in batch.
    l_r_callback: bool, default=False
        set this flag to True in order to enable an increasing learning rate at each epoch.
    Returns
    -------
    model: sklearn model object.
        Fitted model object
    
    history: sklearn history object.
    
    """
    # number of training features
    n_features = X_train.shape[1]

    # Instantiate model
    model = Sequential()

    # Prescribed parameters
    # Add first fully connected layer containing n_features neurons
    # the model will take as input arrays of shape (,n_features) and output arrays of shape (,n_features)
    model.add(Dense(n_features, input_dim=n_features))
    model.add(Activation('relu'))  # rectified linear unit activation
    # add a dropout layer
    model.add(Dropout(0.2))
    # Add second layer, with 12 neurons and ReLU activation
    model.add(Dense(12))
    model.add(Activation('relu'))
    # add a dropout layer
    model.add(Dropout(0.2))
    # Add third layer, with 6 neurons and ReLU activation
    model.add(Dense(6))
    model.add(Activation('relu'))
    # add a dropout layer
    model.add(Dropout(0.2))
    # Last, output layer contains one neuron (ie, target is a real numeric value)
    model.add(Dense(1))
    model.add(Activation('sigmoid'))
    
    # Model Compiling
    model.compile(optimizer=tf.keras.optimizers.Adam(lr=0.001), loss='binary_crossentropy',
                  metrics=['acc',f1_m,precision_m, recall_m])
    
    print("------------------------")
    print(str('DNN model with custom parameters'))
    print("------------------------")
    # list some properties of the network
    print(model.summary())
    
    # Create a learning rate scheduler callback
    # traverse a set of learning rate 
    if l_r_callback:
        lr_scheduler = tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-4 * 10**(epoch/20)) 
        # fit the model
        history = model.fit(X_train, y_train, validation_data=(X_test, y_test), 
                            epochs=epochs, batch_size=batch_size, callbacks=[lr_scheduler])
    else:
        # fit the model
        history = model.fit(X_train, y_train, validation_data=(X_test, y_test), 
                            epochs=epochs, batch_size=batch_size)
            
    return model, history
```

```python
dnn_model, history1 = dnn_func( X_train_scaled, y_train, X_test_scaled, y_test, 
                               epochs=100, batch_size=16, l_r_callback=True)
```

The optimal learning rate is ~$10^{-3}$, i.e. where the Loss-Learning Rate curve approaches its minimum. I use this from now on (which is also the default value).

Time to tune some hyperparameters. Talos is a framework which allows to finetune the hyperparameters specified in a parameter dictionary, similarly to what gridsearch does. Talos creates an 'experiment' and optimizes different networks with different combination of parameters. It can optimize also the number of hidden layers and play with different architectures. 

The following architecture I am going to optimize has a single dropout layer after the first hidden layer. I am going to test the following hyperparameters:
- Number of hidden units in the first hidden layer;
- Number of hidden layers;
- Network geometry (I opted for a funnel architecture, i.e. the next hidden layer has less hidden units than the previous layer);
- batch size;
- dropout;
- the optimizer;

```python
# model definition
def MLP_model(X_train, y_train, X_test, y_test, params):
    """
    Trains, predicts, and evaluates performance of a custom dnn built on top of the params dictionary.

    Parameters
    ----------
    X_train : numpy.ndarray
        train dataset.
    y_train : numpy.ndarray
        corresponding train labels.
    X_test : numpy.ndarray
        test dataset.
    y_test : numpy.ndarray
        corresponding test labels.
    param : python dictionary
        python dictionary containing key-value pairs of parameters and their values to be tested.
    
    Returns
    -------
    out: sklearn model object.
        Fitted model object
    
    model: sklearn model object.
        Fitted model object
    
    history: 
    """
    # Model definition
    model = Sequential()
    model.add(Dense(params['first_neuron'], input_dim=X_train.shape[1],
                    activation=params['activation']))

    model.add(Dropout(params['dropout']))

    # Test for number of layers and shapes
    hidden_layers(model, params, 1)

    # Add last layer with sigmoid activation
    model.add(Dense(1))#, activation='sigmoid'))
    model.add(Activation('sigmoid'))

    # compile the model and optimize the optimizer algorithm and learning rate as well.
    model.compile(optimizer=params['optimizer'](lr=lr_normalizer(params['lr'], params['optimizer'])),
                  loss=params['losses'], metrics=[recall_m])

    # fit the model and optimize the batch size as well.
    out = model.fit(X_train, y_train,
                    validation_data=(X_test, y_test),
                    verbose=0, batch_size=params['batch_size'],
                    epochs=params['epochs'])
    # return history object and model
    return out, model

```
